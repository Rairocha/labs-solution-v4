```python
def indentify_outliers(df,column,limit):
    IQR= df[column].quantile(.75)-df[column].quantile(.25)
    upper_limit = df[column].quantile(.75)+limit*IQR
    return df.loc[df[column] > upper_limit].index
    
```

## - Encoding Categorical Data.

```python
def generate_encoder(df):
    """Trains encoder to change categorical to numerical"""
    from sklearn.preprocessing import OneHotEncoder
    encoder  = OneHotEncoder(drop='if_binary',sparse_output=False)

    encoder.fit(df[['state','coverage','education','employmentstatus','location_code','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class','vehicle_size','response','gender']])
    return encoder

def preprocessing(df,encoder):
    """treats all the categorical features"""
    df['day'] = pd.to_datetime(df['effective_to_date']).dt.day
    df['month'] = pd.to_datetime(df['effective_to_date']).dt.month
    df['year'] = pd.to_datetime(df['effective_to_date']).dt.year
    df = df.drop(columns = ['effective_to_date'])
    
    categorical_columns = ['state','coverage','education','employmentstatus','location_code','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class','vehicle_size','response','gender']
    df_categorical = pd.DataFrame(encoder.transform(df[categorical_columns])
            ,columns=encoder.get_feature_names_out())
    df = df.drop(columns= categorical_columns)
    
       
    df = df.drop(columns= ['customer'])
    
    return pd.concat([df,df_categorical],axis=1)
```

## train test split

```python
def identify_outliers(df,column,limit):
    IQR= df[column].quantile(.75)-df[column].quantile(.25)
    upper_limit = df[column].quantile(.75)+limit*IQR
    return df.loc[df[column] > upper_limit].index

def normalization_step(df,scaler):
    return pd.DataFrame(scaler.transform(df.select_dtypes('number')),columns = df.select_dtypes('number').columns)

def generate_encoder(df):
    """Trains encoder to change categorical to numerical"""
    from sklearn.preprocessing import OneHotEncoder
    encoder  = OneHotEncoder(drop='if_binary',sparse_output=False)

    encoder.fit(df[['state','coverage','education','employmentstatus','location_code','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class','vehicle_size','response','gender']])
    return encoder

def preprocessing(df,encoder):
    """treats all the categorical features"""
    df['day'] = pd.to_datetime(df['effective_to_date']).dt.day
    df['month'] = pd.to_datetime(df['effective_to_date']).dt.month
    df['year'] = pd.to_datetime(df['effective_to_date']).dt.year
    df = df.drop(columns = ['effective_to_date'])
    
    categorical_columns = ['state','coverage','education','employmentstatus','location_code','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class','vehicle_size','response','gender']
    df_categorical = pd.DataFrame(encoder.transform(df[categorical_columns])
            ,columns=encoder.get_feature_names_out(),index=df.index)
    df = df.drop(columns= categorical_columns)
       
    df = df.drop(columns= ['customer'])
    
    return pd.concat([df,df_categorical],axis=1)
```

```python
def train(X_train,y_train):
    from sklearn.preprocessing import StandardScaler
    rows_to_delete = identify_outliers(pd.DataFrame(y_train),'total_claim_amount',4.5)
    X_train = X_train.drop(rows_to_delete)
    y_train = y_train.drop(rows_to_delete)
    
    rows_to_delete = identify_outliers(X_train,'monthly_premium_auto',3.5)
    X_train = X_train.drop(rows_to_delete)
    y_train = y_train.drop(rows_to_delete)
    
    rows_to_delete = identify_outliers(X_train,'customer_lifetime_value',4.5)
    X_train = X_train.drop(rows_to_delete)
    y_train = y_train.drop(rows_to_delete)
    
    encoder = generate_encoder(X_train)
    
    X_train_no_cat = preprocessing(X_train,encoder)
 
    
    scaler= StandardScaler()
    scaler.fit(X_train_no_cat)
    X_train_norm = normalization_step(X_train_no_cat,scaler)
    return X_train_norm,y_train, encoder, scaler
```

```python
def test(X_test,encoder,scaler):
    X_test_no_cat = preprocessing(X_test,encoder)
    X_test_norm =normalization_step(X_test_no_cat,scaler)
    return X_test_norm
```

```python
from sklearn.model_selection import train_test_split
X= df.drop(columns = 'total_claim_amount')
y= df['total_claim_amount']
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
X_train,y_train,encoder,scaler = train(X_train,y_train)
```

```python
X_test = test(X_test,encoder,scaler)
```

# 05 - Modeling

- Apply model.




```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(X_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model_gbr = GradientBoostingRegressor(max_depth=7,n_estimators=30)
model_gbr.fit(X_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor 
model_knn = KNeighborsRegressor(10)
model_knn.fit(X_train,y_train)
```

# 06 - Model Validation

- R2.
- MSE.
- RMSE.
- MAE.


```python
print('R2','Linear Regression',model_lr.score(X_test,y_test),
      'Gradient Boosting',model_gbr.score(X_test,y_test),
      "KNN",model_knn.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('MSE','Linear Regression',mean_squared_error(y_test,model_lr.predict(X_test)),'Gradient Boosting',mean_squared_error(y_test,model_gbr.predict(X_test))
      ,"KNN",mean_squared_error(y_test,model_knn.predict(X_test)))
```

```python
print('RMSE','Linear Regression',mean_squared_error(y_test,model_lr.predict(X_test),squared=False),
      'Gradient Boosting',mean_squared_error(y_test,model_gbr.predict(X_test),squared=False)
      ,"KNN",mean_squared_error(y_test,model_knn.predict(X_test),squared=False))
```

```python
print('MAE','Linear Regression',mean_absolute_error(y_test,model_lr.predict(X_test)),
      'Gradient Boosting',mean_absolute_error(y_test,model_gbr.predict(X_test))
      ,"KNN",mean_absolute_error(y_test,model_knn.predict(X_test)))
```
